```python
# States to consider for the analysis
states = ['Andhra_Pradesh', 'Arunachal_Pradesh', 'Assam', 'Bihar',
       'Chhattisgarh', 'Goa', 'Gujarat', 'Haryana', 'Himachal_Pradesh',
       'Jammu & Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Madhya_Pradesh',
       'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha',
       'Punjab', 'Rajasthan', 'Sikkim', 'Tamil_Nadu', 'Telangana', 'Tripura',
       'Uttar_Pradesh', 'Uttarakhand', 'West_Bengal']
```

### Reading the GDP of individual states. Taking 2014-15 as the base year since the data for all the states(except WB) is present for this year. 

```python
import os, fnmatch

# Defining an empty dataframe. Will add the data of individual state to this dataframe. 
# GSVA = Gross State Value Added
gsva = pd.DataFrame()
# List of files where GSDP files are present
path=("C://Users//kadam//OneDrive//Documents//Upgrad//1. Statistics & EDA//GDP-Assignment//Part1-B//")
listOfFiles = os.listdir(path)  


for filename in listOfFiles:  
    for statename in states:
        if statename in filename:
            filepath = path+filename
            
            # use the following encoding "ISO-8859-1" otherwise it may create problem reading data for some states
            data_ = pd.read_csv(filepath, encoding = "ISO-8859-1")
            # select only the following columns, Items and year
            data_= data_[['Item', '2014-15']]
            # Take the transpose of the data before adding it to the dataframe
            data_ = data_.T
            header=data_.iloc[0]
            data_=data_[1:]
            data_.columns=header
            # print(statename)
            
            # Add data to the dataframe with their respective name of state. 
            # Careful with state having length of 2. Using space bewteen the words of a state as standard as it appears
            # in this format in the dataframe of Part I-A. 
            if statename=='Uttar_Pradesh':
                data_['state'] = 'Uttar Pradesh'
            elif statename=='Madhya_Pradesh':
                data_['state'] = 'Madhya Pradesh'
            elif statename=='Andhra_Pradesh':
                data_['state'] = 'Andhra Pradesh'   
            elif statename=='Arunachal_Pradesh':
                data_['state'] = 'Arunachal Pradesh'   
            elif statename=='Tamil_Nadu':
                data_['state'] = 'Tamil Nadu'   
            else:      
                data_['state'] = statename

            frames = [data_,gsva]
            gsva = pd.concat(frames)
        
        
            
```

```python
gsva.set_index('state', inplace=True)
gsva
```

## Finding the depedency of GDP on primary, secondary and tertiary.

```python
# creating a time slot for various time period of day
def gdp_percentage(gsdp, total_gsdp):
    return gsdp/total_gsdp
    
gsva['primary_gdp_percent'] = gsva[['Primary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
gsva['secondary_gdp_percent'] = gsva[['Secondary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
gsva['tertiary_gdp_percent'] = gsva[['Tertiary','Gross State Domestic Product']].apply(lambda x: gdp_percentage(*x), axis=1)
```

```python
gsva
```

### Plotting the percentage contribution of primary, secondary and tertiary sectors as a percentage of total GSDP for all the states.

```python
gsva = gsva.sort_values('Per Capita GSDP (Rs.)')
gsva[['primary_gdp_percent', 'secondary_gdp_percent', 'tertiary_gdp_percent']].plot(kind='bar', stacked=True)
gsva[["Population ('00)"]].plot(kind='bar')
```

### Some Observations
- For states which has low per capita income, has significant dependence on primary source whereas states which have high per capita income relies more on secondary and tertiary.
- UP has high GDP but due to its population, has low per capita income
- Madhya Pradesh also has low per capita income, but mainly due to over dependence on primary source. 
- Arunachal Pradesh has decent per capita income inspite of heavily dependence on primary source as its population is significantly low.
- States like Assam and Jharkhand go together with similar per capita income, GDP and dependence on primary, secondary and tertiary source.

##  Recommendations

Category I: 'Goa', 'Sikkim', 'Kerala', 'Haryana'. 
There are only 4 states in this category, all of which are small (area wise) and have good industial base, agricultural, trade and real state. For North-Eastern states, Sikkim can be a role model of development. 
- Trade and hospitality bussiness should be promoted further. 
- Manufacturing should be further promoted. 

Category II: 'Arunachal Pradesh', 'Punjab','Telangana', 'Gujarat', 'Karnataka', 'Tamil Nadu','Maharashtra', 'Uttarakhand'
- These states have relatively good per capita gsdp and are driven by around equal percentage by industries and agriculture. This implies that the these states already have pool of talent to drive industries and agriculture. Because of the talent, other industies can take their advantage and set up good industries. 
- States like Karnataka, Tamil Nadu, Gujarat and Maharashtra, which has huge costal area can set up ports. Costal area can help in harness clean energy and fisheries can be increased. Similary, tourism industy should also be encouraged. 

Category III:  'Madhya Pradesh', 'Odisha', 'Meghalaya', 'Tripura', 'Rajasthan', 'Chhattisgarh', 'Nagaland', 'Andhra Pradesh'
- These states are mostly dependent on agriculture and have less number of industries. More programs for educating the youth and promoted which can lead to futher developement of industries. 
- As these states are driven by agriculture, more investment should be done on modern ways of agriculture to increase the productivity of the crops. 
- For states like Andhra Pradesh and Odissa, ports should be developed for the transport of goods which will create new source of employement and also will increase the trade for produced goods. 
- North Indian states like Meghalaya, Tripura and Nagaland  lies in this category. As these states are also driven by agriculture and their population density is also less, which implies points to the lack of good infrastructure for their produced goods. Steps must be taken to connect these states to the whole of India and more industries must be set up. 

Category IV: 'Bihar', 'Uttar Pradesh', 'Manipur' 'Assam', 'Jharkhand'. 
Most of the population in these are dependent on agriculture and manufacturing lies at the 6th category. Since the economy is dependent on agriculture, recommendation are:
- Increase the productivity of the crops by being less dependent on rainfall and more on water sources like canal or underground water. Also, better infrastructure must be developed to preserve the agricultural products for long duration of time.
- Better infrastructure in terms of connectivity must be developed to transport the goods and products from agriculture. If better infrastructure is developed to store the goods, farmers can get better prices for their crops as they do not need to sell their products at cheaper prices if the productivity is high. 
- Industries should be set up for these states. This in turn will help the population to be less dependent on agriculture and help to generate a steady income throughout the year. There should be gradual shift from agriculture to industries. 
- States like Bihar, Uttar Pradesh, Assam have rivers like Ganga, Yamuna, Narmada and Brahmaputra which can be used for cheaper water transport.

- For all the 3 categories(primary, upper primary and secondary), it can be observed that there is almost linear relationship between drop-out rate and per capita GSDP.
